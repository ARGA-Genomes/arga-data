```python
import pandas as pd
import pandas_read_xml as pdx
from pandas_read_xml import flatten, fully_flatten, auto_separate_tables
taxon = 'all'

df = pdx.read_xml(f"/data/arga-data/bold-australia-{taxon}.xml", ['bold_records'], encoding="ISO-8859-1")
# df = pd.read_xml(f"http://v3.boldsystems.org/index.php/API_Public/combined?format=xml&geo=Australia&taxon={taxon}", dtype={'specimen_identifiers': object}) 
# df = pd.read_table("../data/bold_data.tsv", dtype='object', parse_dates=['collectiondate']) 
# columns:
# processid	sampleid	recordID	catalognum	fieldnum	institution_storing	bin_uri	phylum_taxID	phylum_name	
# class_taxID	class_name	order_taxID	order_name	family_taxID	family_name	subfamily_taxID	subfamily_name	
# genus_taxID	genus_name	species_taxID	species_name	subspecies_taxID	subspecies_name	identification_provided_by	
# voucher_type	tissue_type	collectors	collectiondate	lifestage	sex	reproduction	extrainfo	notes	lat	lon	
# coord_source	coord_accuracy	country	province	region	exactsite
df.info()
df
```

```python
df = df.pipe(flatten)
df.info()
```

```python
# df = df.pipe(flatten)
df = df.drop('@xsi:noNamespaceSchemaLocation', axis=1)
df = df.drop('@xmlns:xsi', axis=1)
df.info()
```

```python
df = df.pipe(fully_flatten)
df.info()
df.head(5)
```

```python
# Cleanup field names and map to DwC where possible
field_mapping = {
    # DwC fields
    'record|collection_event|collectiondate': 'eventDate',
    'record|collection_event|collectors': 'recordedBy',
    'record|collection_event|coordinates|coord_accuracy':'coordinatePrecision',
    'record|collection_event|coordinates|coord_source':'georeferenceSources',
    'record|collection_event|coordinates|lat':'decimalLatitude',
    'record|collection_event|coordinates|lon':'decimalLongitude',
    'record|collection_event|country':'country',
    'record|collection_event|exactsite':'locationRemarks',
    'record|collection_event|province':'stateProvince',
    'record|collection_event|region':'locality',
    'record|notes':'fieldNotes', # eventRemarks
    'record|processid':'recordNumber',
    'record|record_id':'occurrenceID',
    'record|bin_uri':'bold_bin_uri',
    # use df.apply to merge following fields into associatedSequences
    # 'record|sequences|sequence|genbank_accession':'ncbi_accession',
    # 'record|sequences|sequence|sequenceID':'',
    # 'record|sequences|sequence|nucleotides':'',
    'record|specimen_details|extrainfo':'identificationRemarks',
    'record|specimen_details|lifestage':'lifeStage',
    'record|specimen_details|reproduction':'bold_reproduction',
    'record|specimen_details|sex':'sex',
    'record|specimen_details|voucher_desc':'dispositionProperty',
    'record|specimen_details|voucher_type':'bold_voucher_type',
    'record|specimen_details|tissue_type': 'preparations',
    'record|specimen_identifiers|catalognum':'otherCatalogNumbers',
    'record|specimen_identifiers|fieldnum':'fieldNumber',
    'record|specimen_identifiers|institution_storing':'ownerInstitutionCode',
    'record|specimen_identifiers|sampleid':'materialSampleID',
    # use df.apply to merge following fields into associatedMedia 
    # specimen_imagery|media|caption 
    # specimen_imagery|media|copyright  
    # specimen_imagery|media|image_file         
    # specimen_imagery|media|mediaID
    # specimen_imagery|media|metatags
    'record|taxonomy|identification_provided_by':'identifiedBy',
    'record|taxonomy|class|taxon|name':'class',
    'record|taxonomy|class|taxon|taxID':'classID',
    'record|taxonomy|family|taxon|name':'family',
    'record|taxonomy|family|taxon|taxID':'familyID',
    'record|taxonomy|genus|taxon|name':'genus',
    'record|taxonomy|genus|taxon|taxID':'genusID',
    'record|taxonomy|order|taxon|name':'order',
    'record|taxonomy|order|taxon|taxID':'orderID',
    'record|taxonomy|phylum|taxon|name':'phylum',
    'record|taxonomy|phylum|taxon|taxID':'phylumID',
    'record|taxonomy|species|taxon|name':'species',
    'record|taxonomy|species|taxon|taxID':'speciesID',
    'record|taxonomy|subspecies|taxon|name':'subspecies',
    'record|taxonomy|subspecies|taxon|taxID':'subspeciesID',
    # use df.apply to merge following fields into associatedMedia 
    # record|tracefiles|read|*
}

# keep copies of DwC matched columns (just in case)
df['scientificName'] = df['record|taxonomy|species|taxon|name'] 

# do rename
df.rename(
  columns = field_mapping, 
  inplace = True
)

df.info()
df.head()
```

```python
# process sequences data
sequenceFields = {
  'sequenceID': 'record|sequences|sequence|sequenceID',
  'genbank_accession': 'record|sequences|sequence|genbank_accession',
  'markercode': 'record|sequences|sequence|markercode',
  'nucleotides': 'record|sequences|sequence|nucleotides',
}

def addAssociatedeSequenceData(row):
  sequenceData = {}
  for sd in sequenceFields:
    sequenceData[sd] = row[sequenceFields[sd]]
  return [sequenceData]

df['associatedSequences'] = df.apply(addAssociatedeSequenceData, axis=1)
df.drop(sequenceFields.values(), axis=1, inplace=True)

# process images data
imagesFields = {
  'mediaID': 'record|specimen_imagery|media|mediaID',
  'image_file': 'record|specimen_imagery|media|image_file',
  'metatags': 'record|specimen_imagery|media|metatags',
  'copyright': 'record|specimen_imagery|media|copyright',
  'caption': 'record|specimen_imagery|media|caption',
}

def associatedImageData(row):
  imageData = {}
  for id in imagesFields:
    imageData[id] = row[imagesFields[id]]
  return [imageData]

df['associatedMedia'] = df.apply(associatedImageData, axis=1)
df.drop(imagesFields.values(), axis=1, inplace=True)

# process traces data
tracesFields = {
  'direction': 'record|tracefiles|read|direction',
  'read_id': 'record|tracefiles|read|read_id',
  'markercode': 'record|tracefiles|read|markercode',
  'run_date': 'record|tracefiles|read|run_date',
  'seq_primer': 'record|tracefiles|read|seq_primer',
  'sequencing_center': 'record|tracefiles|read|sequencing_center',
  'trace_link': 'record|tracefiles|read|trace_link',
}

def addTracesData(row):
  tdata = {}
  for tf in tracesFields:
    tdata[tf] = row[tracesFields[tf]]
  return [tdata]

df['bold_tracefiles_read'] = df.apply(addTracesData, axis=1)
df.drop(tracesFields.values(), axis=1, inplace=True)
  
df.drop(['record|taxonomy|subfamily|taxon|name','record|taxonomy|subfamily|taxon|taxID'], axis=1, inplace=True)
df.info()
df.head()
```

```python
# Remove duplicates (array values in some fields) 
# TODO parse these properly into an array as per DwC spec for associatedSequences and associatedMedia
df = df.drop_duplicates(subset=['occurrenceID'],keep='first')
df.info()
```

```python
# save to CSV file
df.to_csv(f"/data/arga-data/bold_australia_{taxon}_data.csv", index=False)
```
